```python
import pandas as pd
```

```python
# o "encoding='latin1'" é para informar que o arquivo é codificado por 'latin1'.

# o "sep=';'" é para informar que o arquivo é separado por ponto e virgula.

# o "low_memory=False" é para não dar erro.

# "decimal=','" é para informar que os números são separados por vírgula e não por ponto.
```

```python
precos_df = pd.read_csv('TA_PRECO_MEDICAMENTO.csv', encoding='iso-8859-1', sep=';', low_memory=False)
display(precos_df)
```

```python
# laço for para visualizar o nome de todas as colunas
for col in precos_df.columns:
    print(col)
```

```python
# Pesquisando pela palavra inteira do medicamento

# localizando medicamento 'SINGULAIR' e mostrando as colunas 'PRODUTO', 'APRESENTAÇÃO', 'COMERCIALIZAÇÃO 2020'(só desse ano) 
# e 'PF Sem Impostos'

precos_medic_df = precos_df.loc[(precos_df['PRODUTO'] == 'SINGULAIR') & (precos_df['COMERCIALIZAÇÃO 2020'] == 'Sim'), 
                                ['PRODUTO', 'APRESENTAÇÃO', 'COMERCIALIZAÇÃO 2020', 'PF Sem Impostos']]
```

```python
display(precos_medic_df)
```

```python
# Pesquisando pela palavra incompleta do medicamento

# o 'match' é usado para poder pesquisar algum elemento que contenha alguma parte da palavra ou frase
precos_medic_df = precos_df.loc[(precos_df['PRODUTO'].str.match('SINGUL')) & (precos_df['COMERCIALIZAÇÃO 2020'] == 'Sim'), 
                                ['PRODUTO', 'APRESENTAÇÃO', 'COMERCIALIZAÇÃO 2020', 'PF Sem Impostos']]
```

```python
display(precos_medic_df)
```

```python
# procurando por código de barras
precos_codigo_df = precos_df.loc[precos_df['EAN 1'] == '7897337706575']
```

```python
display(precos_codigo_df)
```

```python
# mostrando a coluna PMC 0% do produto 'SINGULAIR' do EAN1 '7897337706575'

pmc_df = precos_df.loc[precos_df['PRODUTO'] == 'SINGULAIR', ['PMC 0%']]
display(pmc_df)
```

```python
# mostrar a coluna '(PIS/COFINS)' apenas nas comercializações em 2020:

lista_df = precos_df.loc[(precos_df['COMERCIALIZAÇÃO 2020'] == 'Sim') & (precos_df['LISTA DE CONCESSÃO DE CRÉDITO TRIBUTÁRIO (PIS/COFINS)']), 
                                ['COMERCIALIZAÇÃO 2020', 'LISTA DE CONCESSÃO DE CRÉDITO TRIBUTÁRIO (PIS/COFINS)']]
display(lista_df)
```
